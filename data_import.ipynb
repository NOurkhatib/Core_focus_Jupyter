```python
import requests
import io
import PyPDF2
```

```python
def get_request_query(query=""):
    api_url = "https://gegevensmagazijn.tweedekamer.nl/OData/v4/2.0/"
    query = api_url + str(query)

    response = requests.get(query)
    return response
```

```python
def check_response(response):
     # Che  ck if the request was successful
    if response.status_code == 200:
        print("Headers:", response.headers)
        try:
            data = response.json()
            
            # print("Data structure:", data.keys())
            return data['value']
        except ValueError:
            print("Response content is not in JSON format.")
            print(response.text)
    else:
        print("Error:", response.status_code)
        print(response.text)
```

```python
def get_pdf(response):
    # Assuming 'response' is the response object from requests.get
    if response.status_code == 200:
        # Dictionary to hold page numbers and their text
        pages_text = {} 

        try:
            # Use io.BytesIO to handle the binary stream
            with io.BytesIO(response.content) as open_pdf_file:
                # Read the PDF
                pdf_reader = PyPDF2.PdfReader(open_pdf_file)
                text = ''
                for page_num in range(len(pdf_reader.pages)):
                    # # Extract text from each page
                    # text += pdf_reader.pages[page_num].extract_text()

                    # Extract text from each page
                    page_text = pdf_reader.pages[page_num].extract_text()
                    # Store the text in the dictionary
                    pages_text[page_num + 1] = page_text

                # print(text)  # The 'text' variable contains the extracted text
                return pages_text
        except Exception as e:
            print("An error occurred:", e)
    else:
        print("Failed to retrieve PDF. Status code:", response.status_code)

        
```
